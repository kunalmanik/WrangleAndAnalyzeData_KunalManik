<h3>Twitter API</h3>

```python
CONSUMER_KEY = ''
CONSUMER_SECRET = ''
ACCESS_TOKEN = ''
ACCESS_SECRET = ''

auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_SECRET)

try:
    api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
except:
    print("ERROR")
```

```python
# Store data on tweet_json.txt
with open('tweet_json.txt', 'w') as outfile: 
    for tweet_id in df_twitter_archive['tweet_id']:
        pre_start = time.time()
        try:
            tweet = api.get_status(tweet_id)
            # Convert to JSON and write
            start = time.time()
            json.dump(tweet._json, outfile)
            outfile.write("\n")
            end = time.time()
            print('Elapsed time for tweet id {} : {}'.format(tweet_id, end - start))            
            
        except:  
            # In case the tweet does not exist anymore
            print('ERROR: Error on tweet id {}'.format(tweet_id))
            e = os.sys.exc_info()[0]
            
        post_end = time.time()
        print('Total time Elapsed : {}'.format(post_end - pre_start))
```

```python
final_columns = ['id', 'favorite_count', 'retweet_count']
data = []

with open('tweet_json.txt', 'r') as read_json_tweet: 
    
    tweet_json = read_json_tweet.readline()
    
    while tweet_json:
        tweet_dict = json.loads(tweet_json)
        
        data_row = dict((k, tweet_dict[k]) for k in final_columns)
        data.append(data_row)
        
        tweet_json = read_json_tweet.readline()
        
df_tweet_info = pd.DataFrame.from_dict(data)
```

```python
elements_to_save = ['id', 'favorite_count', 'retweet_count']
data = []

with open('tweet_json.txt', 'r') as read_json_tweet:     
    tweet_json = read_json_tweet.readline()
    
    while tweet_json:
        tweet_dict = json.loads(tweet_json)
        
        data_row = dict((k, tweet_dict[k]) for k in elements_to_save)
        data.append(data_row)
        
        tweet_json = read_json_tweet.readline()
        
df_tweet_info = pd.DataFrame.from_dict(data)
```

<h3>Cleaning Data</h3>

```python
# Making copies
df_clean_twitter_archive = df_twitter_archive.copy()
df_clean_image_predictions = df_image_predictions.copy()
df_clean_tweet_info = df_tweet_info.copy()
```

**There are many names in the dataset that are not actual names.**

```python
def rectify_name(row):    
    if row['name'] == 'a':            
        text = row['text']
        name_sub_list = text.split()
        
        name = 'None'

        if 'named' in name_sub_list:
            name_pos = name_sub_list.index('named') + 1
            name = name_sub_list[name_pos]
        
        elif 'name is' in text:            
            name_pos = name_sub_list.index('name') + 2
            name = name_sub_list[name_pos]
        
        row['name'] = name.replace('.','')
    
    return row
    
df_clean_twitter_archive = df_clean_twitter_archive.apply(rectify_name, axis=1)
```

```python
'a' in df_clean_twitter_archive['name']
```

```python
# Rectifying datatypes

df_clean_twitter_archive.timestamp = pd.to_datetime(df_clean_twitter_archive.timestamp)
df_clean_twitter_archive.retweeted_status_timestamp = pd.to_datetime(df_clean_twitter_archive.retweeted_status_timestamp)
```

```python
df_clean_twitter_archive.info()
```

```python
# Expanding url - in case it's not null

def expand_urls(row):    
    
    if pd.notnull(row['expanded_urls']):
        return row
    
    else:        
        tweet_id = row['tweet_id']        
        row['expanded_urls'] = 'https://twitter.com/dog_rates/status/{}'.format(tweet_id)
        
        return row

df_clean_twitter_archive = df_clean_twitter_archive.apply(expand_urls, axis=1)
```

```python
df_clean_twitter_archive[df_clean_twitter_archive['expanded_urls'].isnull()]
```

```python
# Rectifying incorrect ratings

incorrect_indices = [x-1 for x in [1069, 1166, 2336]]
incorrect_rows = df_clean_twitter_archive.iloc[incorrect_indices]

for txt in incorrect_rows.text:
    print(txt)

df_clean_twitter_archive.at[incorrect_indices[0], 'rating_numerator'] = 14
df_clean_twitter_archive.at[incorrect_indices[0], 'rating_denominator'] = 10

df_clean_twitter_archive.at[incorrect_indices[1], 'rating_numerator'] = 13
df_clean_twitter_archive.at[incorrect_indices[1], 'rating_denominator'] = 10

df_clean_twitter_archive.at[incorrect_indices[2], 'rating_numerator'] = 9
df_clean_twitter_archive.at[incorrect_indices[2], 'rating_denominator'] = 10
```

```python
incorrect_rows = df_clean_twitter_archive.iloc[incorrect_indices]

values = zip(incorrect_rows.text, incorrect_rows.rating_numerator, incorrect_rows.rating_denominator)
for (text, numerator, denominator) in values:
        print('{}/{} from: \n\t"{}"'.format(numerator, denominator, text))
```

```python
# Removing 343rd entry since it's not related
df_clean_twitter_archive = df_clean_twitter_archive.drop(df_clean_twitter_archive.index[342])
```

```python
# Rectifying missing values - "puppers"

def add_puppers(row):

    if 'puppers' in row['text']:
        row['pupper'] = 'pupper'
        
    return row

df_clean_twitter_archive = df_clean_twitter_archive.apply(add_puppers, axis=1)
```

```python
# Rectifying Retweets

retweets = df_clean_twitter_archive.text.str.match('^RT @')

df_clean_twitter_archive = df_clean_twitter_archive[~retweets]
```

```python
retweets = df_clean_twitter_archive.text.str.match('^RT @')

df_clean_twitter_archive[retweets]
```

```python
# Getting final rating in the form of decimals

def get_final_rating(row):
    #if text matches for numerator, change it
    decimal_numerator_pattern = '(\d+\.\d+)\/(\d+)'
    matches = re.search(decimal_numerator_pattern, row['text'])
    if matches != None:
        row['rating_numerator'] = matches.group(1)
        row['rating_denominator'] = matches.group(2)

    return row
        
df_clean_twitter_archive = df_clean_twitter_archive.apply(get_final_rating, axis=1)
```

```python
# Rectifying text

df_clean_image_predictions['p1'] = df_clean_image_predictions['p1'].str.lower()
df_clean_image_predictions['p2'] = df_clean_image_predictions['p2'].str.lower()
df_clean_image_predictions['p3'] = df_clean_image_predictions['p3'].str.lower()
```

```python
# Handling missing data in "df_clean_twitter_archive" and "df_clean_image_predictions"

df1 = df_clean_twitter_archive
df2 = df_clean_image_predictions

df_clean_final = pd.merge(df1, df2, how='left', on=['tweet_id'])
```

```python
tweet_ids = df_clean_image_predictions.tweet_id

for tweet_id in tweet_ids:
    
    if tweet_id not in df_clean_twitter_archive.tweet_id.values:
        if tweet_id in df_clean_final.tweet_id.values:
            print('ERROR while cleaning!')
```

```python
df1 = df_clean_final
df2 = df_clean_tweet_info
df2.rename(columns={'id':'tweet_id'}, inplace=True)

df_clean_final = pd.merge(df1, df2, how='left', on=['tweet_id'])
```

```python
tweet_ids = df_clean_tweet_info.tweet_id

for tweet_id in tweet_ids:

    if tweet_id not in df_clean_twitter_archive.tweet_id.values:
        if tweet_id in df_clean_final.tweet_id.values:
            print('ERROR while cleaning!')
            
df_clean_final.info()
```

```python
# Rectifying dog_stage

df_clean_final['dog_stage'] = 'None'

def get_dog_stage(row):
    
    dog_stage = []
    
    if row['doggo'] == 'doggo':
        dog_stage.append('doggo')
    if row['floofer'] == 'floofer':
        dog_stage.append('floofer')
    if row['pupper'] == 'pupper':
        dog_stage.append('pupper')
    if row['puppo'] == 'puppo':
        dog_stage.append('puppo')
        
    if len(dog_stage) < 1: 
        row['dog_stage'] = 'None'
    else:         
        row['dog_stage'] = ','.join(dog_stage)
    
    return row

df_clean_final = df_clean_final.apply(get_dog_stage, axis=1)

# Dropping redundant columns
df_clean_final = df_clean_final.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis=1)
```

```python
# Dropping retweet columns
retweet_cols = ['retweeted_status_id', 'retweeted_status_user_id', 'retweeted_status_timestamp']
df_clean_final = df_clean_final.drop(retweet_cols, axis=1)
```

<h4>Favorites vs Retweets</h4>

Observing how Retweets relate to the number of favorites on a tweet depending on the stage the dog is at.

```python
df = df_clean_final[['retweet_count', 'favorite_count', 'dog_stage']]

y, x = df.retweet_count, df.favorite_count
mask = ~np.isnan(x) & ~np.isnan(y)
slope, intercept, r_value, p_value, std_err = stats.linregress(x[mask], y[mask])
line = slope*x+intercept
plt.xlabel('favorite_count')
plt.ylabel('retweet_count')
plt.title('Favorites vs Retweets')
plt.plot(x, y, 'o', x, line, color='Black', lw='1')

print('Line Equation: retweets = ({})*favorites + {}'.format(slope,intercept))
print('Correlation of line of best fit: {}'.format(r_value))

df_puppers = df[df['dog_stage'] == 'pupper']
df_puppo = df[df['dog_stage'] == 'puppo']
df_doggo = df[df['dog_stage'] == 'doggo']
df_floofer = df[df['dog_stage'] == 'floofer']
df_none = df[df['dog_stage'] == 'None']


bx = df_puppers.plot(kind = 'scatter', x='favorite_count', y='retweet_count', color='Red', alpha=0.5)
df_doggo.plot(kind = 'scatter', x='favorite_count', y='retweet_count', color='Black', ax=bx, alpha=0.5)
df_puppo.plot(kind = 'scatter', x='favorite_count', y='retweet_count', color='Blue', ax=bx, alpha=0.5)
df_floofer.plot(kind = 'scatter', x='favorite_count', y='retweet_count', color='Green', ax=bx, alpha=0.5)


# Legend
puppers = mpatches.Patch(label='Pupper', color = 'Red')
doggo = mpatches.Patch(label='Doggo', color = 'Black')
puppo = mpatches.Patch(label='Puppo', color = 'Blue')
floofer = mpatches.Patch(label='Floofer', color = 'Green')

plt.legend(handles=[puppers, doggo, puppo, floofer])
plt.title('Favorites vs Retweets (Grouped by Dog Stages)')
plt.plot(x, line, color='gray', lw='1')

plt.show();
```

I plotted a scatterplot of all tweets.

It can be observed that most of the tweets have less than 40,000 favorites. It can also be observed that from the equation for the line of best fit that there tends to be 3 times as many favorites as retweets. However we notice that as the values get larger, the ratio of retweets to favorites get larger too.

On grouping the plot with the dog stages (for those data is available), smaller values stick-close below the original line of best fit.

```python
# Zooming into the part in the plot where most data is represented
xlim = (0, 25000)
ylim = (0, 10000)

cx = df_puppers.plot(kind = 'scatter', x='favorite_count', y='retweet_count', color='Red', alpha=0.5, ylim=ylim, xlim=xlim)
df_doggo.plot(kind = 'scatter', x='favorite_count', y='retweet_count', color='Black', ax=cx, alpha=0.5, ylim=ylim, xlim=xlim)
df_puppo.plot(kind = 'scatter', x='favorite_count', y='retweet_count', color='Blue', ax=cx, alpha=0.5, ylim=ylim, xlim=xlim)
df_floofer.plot(kind = 'scatter', x='favorite_count', y='retweet_count', color='Green', ax=cx, alpha=0.5, ylim=ylim, xlim=xlim)

plt.legend(handles=[puppers, doggo, puppo, floofer])
plt.title('Favorites vs Retweets (Zoomed)')
plt.plot(x, line, color='gray', lw='1')

plt.show()
```

After zooming in, it can be observed that there are more tweets classified as puppers than any other existing classification. Most of the puppers are lower valued and tend to have a higher ratio of retweets to favorites than the line of best fit. Almost every other classification tends to have a lower ratio of retweets to favorites (more favorites) than the puppers classification.The tweets classified withing doggos tend to have higher values as compared to the pupper tweets.

<h4>Most Common Dog Breeds</h4>

```python
required_attr = ['retweet_count', 'favorite_count', 'rating_denominator', 'rating_numerator']
df_dog_breeds = df_clean_final[required_attr].copy()

def rating_percentage(row):
    if row['rating_denominator'] == 0:
        return 0
    return row['rating_numerator']/row['rating_denominator']


df_dog_breeds['rating'] = df_dog_breeds.apply(rating_percentage, axis=1)
```

```python
required_attr = ['p1', 'p1_dog', 'p1_conf', 'p2', 'p2_dog', 'p2_conf', 'p3', 'p3_dog', 'p3_conf']
df_breed_possibility = df_clean_final[required_attr].copy()

df_breed_possibility['breed'] = ''
df_breed_possibility['breed_conf'] = 1

def nearest_breed(row):
    
    breed = 'not_dog'
    confidence = 0
    
    dog_preds = [(row[['p1', 'p1_dog', 'p1_conf']]), (row[['p2', 'p2_dog', 'p2_conf']]), (row[['p3', 'p3_dog', 'p3_conf']])]
    index_breed, index_isDog, index_conf = 0,1,2
    
    for pred in dog_preds:        
        if pred[index_isDog]:            
            if pred[index_conf] >= confidence:
                breed = pred[index_breed]
                confidence = pred[index_conf]
    
    # Updating breed list
    row['breed'] = breed
    row['breed_conf'] = confidence
    return row
        
    breeds = [row['p1'], row['p2'], row['p3']]
    
df_breed_possibility = df_breed_possibility.apply(nearest_breed, axis=1)

df_dog_breeds['breed'] = df_breed_possibility['breed']

df_dog_breeds['breed_conf'] = df_breed_possibility['breed_conf']
```

```python
# 12 most common dog breeds excluding "not_dog"

only_dogs = df_dog_breeds['breed'] != 'not_dog'
#df_dog_breeds[only_dogs].breed.value_counts()[12::-1].plot(kind='bar');
temp_df = df_dog_breeds[only_dogs].breed.value_counts()[12::-1];

temp_df.plot(kind='barh')
plt.title('12 Most Common Dog Breeds')
plt.show();
```

Initially, 30 most common dog breeds was plotted. It can be observed that nearly 500 tweets were classified as not a dog (not_dog). This is about 3 times more than the highest numbered actual dog breed. This could be that many tweets from this ratings archive are not dogs, however this seems quite strange. It is possible that the image predictions has misclassified many of the images as not dogs. Next,12 most common (actual) dog breeds were plotted. It can be seen that according to this data, the most common dog breed in the tweets was the **Golden retriever** with 158 tweets. The next five most common breeds were the **Labrador Retriever**, **Pembroke**, **Chihuahua**, **Pug** and **Toy Poodle**.

However, concluding too much from that data since must invoke a sense of extra care since, we are relying on the image prediction code to inform us of the dog breed. It is possible that certain dog breeds were harder to predict.
